```python
# Ejemplo de uso de Seaborn para crear un gráfico de conteo que muestra la cantidad de películas
# y programas de TV por tipo, desglosado por género del director. Utilizamos la columna "type"
# en el eje xy la columna "gender director" como el desglose de colores (hue).
sns.countplot(x='type', hue='gender director', data=df)
plt.title('Count of Movies and TV Shows by Gender of Directors')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
# El segundo ejemplo utiliza Matplotlib para crear un gráfico de torta
# que muestra la distribución de género de los directores en forma de porcentaje.
# Usé la columna "gender director" para contar las ocurrencias de cada género
# y generé las etiquetas y tamaños correspondientes.
gender_counts = df['gender director'].value_counts()
labels = gender_counts.index
sizes = gender_counts.values

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Gender Distribution of Directors')
plt.show()
```

```python
# Generará un gráfico de barras que muestra la distribución de género de los directores en diferentes países.
# Cada barra representa la cantidad de directores masculinos y femeninos en un país específico.

# Usé Seaborn para establecer el estilo del gráfico y configuramos el tamaño de la figura.
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))
# Usé para generar un gráfico de barras que muestra la cantidad de directores de cada género en cada país.
sns.countplot(x='director country', hue='gender director', data=df)
# Le puse título, etiquetas de los ejes y una leyenda al gráfico.
plt.title('Distribución de género de los directores en diferentes países')
plt.xlabel('País del director')
plt.ylabel('Cantidad')
plt.legend(title='Género', loc='upper right')
# Roté las etiquetas del eje x para una mejor legibilidad
plt.xticks(rotation=45)
plt.show()
```
